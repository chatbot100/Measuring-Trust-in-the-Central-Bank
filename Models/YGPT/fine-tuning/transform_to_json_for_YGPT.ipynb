```python
import json
import pandas as pd
```

## Датасет для обучения

```python
data = pd.read_excel('marked_data.xlsx')[:7000]
```

```python
classes = ['negative', 'neutral', 'positive']

def row_to_json(row):
    result = {"text": row['text']}
    for c in classes:
        result[c] = 0
    if row['labels'] == -1:
        result['negative'] = 1
    if row['labels'] == 0:
        result['neutral'] = 1
    if row['labels'] == 1:
        result['positive'] = 1
    return result
```

```python
json_data = [row_to_json(row) for _, row in data.iterrows()]
```

```python
with open('train_dataset.json', 'w', encoding='utf-8') as f:
    for item in json_data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')
```

## Датасет для валидации

```python
df = pd.read_excel('marked_data.xlsx')[7000:7500]
```

```python
classes = ['negative', 'neutral', 'positive']

def row_to_json(row):
    if len(row['text'])>7500:
        row['text'] = row['text'][:7500]
    result = {"text": row['text']}
    for c in classes:
        result[c] = 0
    if row['labels'] == -1:
        result['negative'] = 1
    if row['labels'] == 0:
        result['neutral'] = 1
    if row['labels'] == 1:
        result['positive'] = 1
    return result
```

```python
json_df = [row_to_json(row) for _, row in df.iterrows()]
```

```python
with open('val_dataset.json', 'w', encoding='utf-8') as f:
    for item in json_df:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')
```
