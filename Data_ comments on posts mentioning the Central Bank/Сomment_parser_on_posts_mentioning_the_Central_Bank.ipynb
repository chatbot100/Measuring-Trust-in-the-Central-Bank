```python
offset=0
while offset<1700:
        req=requests.get('https://api.vk.com/method/wall.search',
               params={
                   'access_token': token,
                   'v': 5.131,
                   'domain': group_name,
                   'owners_only': 1,
                   'query': '–±–∞–Ω–∫ —Ä–æ—Å—Å–∏–∏',
                   'count': count,
                   'offset': offset
                   
               })
        res=req.json()['response']['items']
        offset+=100
        time.sleep(1)
        all_posts.extend(res)
```

```python
df = pd.json_normalize(all_posts)
df=df[["id", "owner_id", "date", "text", "comments.count"]]
df.rename(columns={'date': 'post_time','text':'post_text'}, inplace=True)
df['bool']=0
for i in range (len(df)):
    s=str(df.iloc[i]['post_text'])
    s=str(s)
    s=s.lower()
    c=bool(re.search(r'–±–∞–Ω–∫ —Ä–æ—Å—Å–∏–∏|–±–∞–Ω–∫–∞ —Ä–æ—Å—Å–∏–∏|–±–∞–Ω–∫—É —Ä–æ—Å—Å–∏–∏|–±–∞–Ω–∫–æ–º —Ä–æ—Å—Å–∏–∏|–±–∞–Ω–∫–µ —Ä–æ—Å—Å–∏–∏|—Ü–±|—Ü–µ–Ω—Ç—Ä–æ–±–∞–Ω–∫|—Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π –±–∞–Ω–∫|—Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–≥–æ –±–∞–Ω–∫–∞|—Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–º—É –±–∞–Ω–∫—É|—Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–º –±–∞–Ω–∫–æ–º|—Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–º –±–∞–Ω–∫–µ', s))
    df.loc[i,'bool']=df['bool'][i]+c
df=df[df['bool']==1]
df
```

```python
posts_id1=list(df[df['comments.count']!=0]['id'])
```
